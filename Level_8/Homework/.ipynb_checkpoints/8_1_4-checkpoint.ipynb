## Matrices

### Exercise 4
Create two 5x5 matrices of numbers:  
a. Multiply both matrices together.  
b. Compute the eigenvalues of the matrices.  
c. Calculate the correlation matrix for each of the two matrices.  

```python
from IPython.core.interactiveshell import InteractiveShell

InteractiveShell.ast_node_interactivity = "all"
```

```python
import numpy as np

# Create NDarrays containing 25 random number
arr1 = np.random.random(25)
arr2 = np.random.random(25)

# Create matrix from previous NDarrays, resize to 5x5 matrix
mat1 = np.resize(np.array(arr1), (5,5))
mat2 = np.resize(np.array(arr2), (5,5))
mat1
mat2
```

#### a. Multiply both matrices together.

```python
# Matrix multiplication
np.matmul(mat1, mat2)
```

#### b. Compute the eigenvalues of the matrices.

```python
# Compute eigenvalues
np.linalg.eig(mat1)
np.linalg.eig(mat2)
```

#### c. Calculate the correlation matrix for each of the two matrices.

```python
np.corrcoef(mat1,mat2)
```
