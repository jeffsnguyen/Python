```python
import pandas as pd
import numpy as np

# Load the CSV
aaplData = pd.read_csv('aapl_data.csv')
aaplData
```

#### a. Display the first five and last five rows.

```python
# First 5 rows
aaplData.head(5)

# Last 5 rows
aaplData.tail(5)
```

#### b. Display a table containing only dates that have a volume of at least 10,000,000.

```python
above_1e7_Vol = aaplData[aaplData.Volume >= 10000000]
above_1e7_Vol.head()
```

#### c. Display the total volume.

```python
np.sum(aaplData.Volume)
```

#### d. Display the average daily volume.

```python
np.mean(aaplData.Volume)
```

#### e. Add a column containing the daily ‘return’

```python
aaplData['Daily Return'] = (aaplData['Close'] - aaplData['Close'].shift(1)) / aaplData['Close'].shift(1)
aaplData
```

#### f. Calculate the average daily return.

```python
np.mean(aaplData['Daily Return'])
```

#### g. Calculate the volume-weighted average daily return.

```python
aaplClean = aaplData.dropna()  # Drop na values and assign to a new DF
np.average(aaplClean['Daily Return'], weights=aaplClean.Volume)
```

#### h. Add a column containing the 1-week (5-day) moving average volume.

```python
aaplData['1-week MA Volume'] = aaplData['Volume'].rolling(5).mean()
aaplData
```

#### i. Add a column containing the 1-week (5-day) moving average daily return.

```python
aaplData['1-week MA Daily Return'] = aaplData['Daily Return'].rolling(5).mean()
aaplData
```
