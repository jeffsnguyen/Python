# Numpy vs. Lists

```python
import numpy as np
from timer import timer, Timer
import random

@timer
def testNP():
    np.random.random(1000000)
    
@timer
def testPyt():
    [random.random() for i in range(1000000)]
```

```python
testNP()
testPyt()
```

```python
@timer
def testNP(data):
    data.sum()

@timer
def testNP2(data):
    sum(data)
    
@timer
def testPyt(data):
    sum(data)
```

```python
testNP(np.random.random(5000000))
testNP2(np.random.random(5000000))
testPyt([random.random() for i in range(5000000)])
```
