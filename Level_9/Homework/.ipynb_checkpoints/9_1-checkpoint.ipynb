## Pandas Plotting

### Exercise 1
Create a DataFrame containing the sin of all 1/10 numbers from 0 to 50. Plot this ‘sine wave’ using Panda’s plot method.

```python
from IPython.core.interactiveshell import InteractiveShell

InteractiveShell.ast_node_interactivity = "all"
```

```python
import numpy as np
import pandas as pd

arr_x = np.arange(0, 50)  # array of x
arr_f = np.sin(np.divide(arr_x, 10))

# Create the df
df = pd.DataFrame({'x': arr_x, 'sin(.1*x)': arr_f}).set_index('x')
df
```

```python
# Plot
df.plot(figsize=(20,5), grid=True, legend=['sin(.1x)'], xlim=[0, 60])
```
